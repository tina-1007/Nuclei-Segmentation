## Perform inference with a MMDet detector
MMDetection already provides high level APIs to do inference and training.

```python
from mmdet.apis import inference_detector, init_detector, show_result_pyplot

# Choose to use a config and initialize the detector
config = 'configs/my_config.py'
# Setup a checkpoint file to load
checkpoint = 'epoch_50.pth'
# initialize the detector
model = init_detector(config, checkpoint, device='cuda:0')
```

```python
# Use the detector to do inference
img = 'data/test/TCGA-50-5931-01Z-00-DX1.png'
result = inference_detector(model, img)
```

```python
# Let's plot the result  # scale16 epoch40
show_result_pyplot(model, img, result, score_thr=0.3)
```
